```python
#Q29. DISPLAYED A SINGLE CODE FOR THE Measures of Dispersion, FOLLOWED BY VISUALISATION.
product_details = ['Value','Weight','Quantity']
def UVA_numeric(data, var_group):
    size = len(var_group)
    plt.figure(figsize=(7 * size, 4), dpi=100)

    for j, i in enumerate(var_group):
        # Check if the column contains numeric data
        if pd.api.types.is_numeric_dtype(data[i]):
            mini = round(data[i].min(), 2)
            maxi = round(data[i].max(), 2)
            ran = round(maxi - mini, 2)
            mean = round(data[i].mean(), 2)
            median = round(data[i].median(), 2)
            st_dev = round(data[i].std(), 2)
            skew = round(data[i].skew(), 2)
            kurt = round(data[i].kurtosis(), 2)

            # Calculating points for the range within one standard deviation from mean
            points = round(mean - st_dev, 2), round(mean + st_dev, 2)

            # Plotting the variable with every information
            plt.subplot(1, size, j + 1)
            sns.kdeplot(x=data[i], fill=True, label='Density')

            # Plot standard deviation range
            sns.lineplot(x=points, y=[0, 0], color='black', label=f"1 std_dev: {points[0]} to {points[1]}")
            
            # Plot min and max points
            sns.scatterplot(x=[mini, maxi], y=[0, 0], color='orange', s=100, label=f"Min/Max: {mini}/{maxi}")

            # Plot mean
            sns.scatterplot(x=[mean], y=[0], color='red', s=100, label=f"Mean: {mean}")

            # Plot median
            sns.scatterplot(x=[median], y=[0], color='blue', s=100, label=f"Median: {median}")

            # Add plot labels and title
            plt.xlabel(f'{i}', fontsize=14)
            plt.ylabel('Density')
            plt.title(f'Range: {ran}, Kurtosis: {kurt}, Skew: {skew}', fontsize=12)
            plt.legend()
        else:
            print(f"Skipping non-numeric column: {i}")

    plt.tight_layout()
    plt.show()
```

```python
UVA_numeric(data,product_details)
```

```python
#41. Find the Confidence Interval (CI) for the Mean of the non Categorial dataset.
mean_value = df['Value'].mean()
std_dev = df['Value'].std()
cv = (std_dev / mean_value) * 100  # Coefficient of Variation in percentage

confidence_level = 0.95
n = len(df['Value'])
z_score = stats.norm.ppf((1 + confidence_level) / 2)  # Z-score for 95% CI
standard_error = std_dev / np.sqrt(n)
margin_of_error = z_score * standard_error
lower_bound = mean_value - margin_of_error
upper_bound = mean_value + margin_of_error
confidence_interval = (lower_bound, upper_bound)
print(f"95% Confidence Interval for the Mean: {confidence_interval}")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Perform ANOVA for each non-categorical variable using 'Import_Export' as the grouping variable
for var in ['Quantity', 'Value', 'Weight']:
    # Ordinary least squares model
    model = ols(f'{var} ~ C(Import_Export)', data=df).fit()
    # ANOVA table
    anova_table = sm.stats.anova_lm(model, typ=2)
    
    # Print ANOVA results for each variable
    print(f"\nANOVA results for {var}:")
    print(anova_table)
```

**How can you compare the variances of two samples using an F-test in Python?**

```python
import numpy as np
from scipy import stats

# Example data: two samples
sample1 = df['Quantity']
sample2 = df['Weight']

# Calculate variances
var1 = np.var(sample1, ddof=1)  # ddof=1 for sample variance
var2 = np.var(sample2, ddof=1)

# Calculate F-statistic
F_statistic = var1 / var2

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# p-value from F-distribution
p_value = stats.f.sf(F_statistic, df1, df2)

print(f'F-statistic: {F_statistic}')
print(f'p-value: {p_value}')

# Compare p-value to significance level (e.g., 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

### Analysis:
The F-statistic of 4.03 and an extremely low p-value (approximately 0) indicate strong evidence against the null hypothesis, suggesting that the variances of **Quantity** and **Weight** are significantly different. This means that the degree of variability in these two variables is not the same, highlighting a potential distinction in how these measures behave across the dataset.

### Managerial Insight:
1. **Understanding Variability**: The significant difference in variances suggests that the factors influencing Quantity and Weight may differ, requiring tailored management strategies for each variable.
2. **Quality Control**: Variability in Quantity may affect supply chain operations; managers should monitor processes to understand and mitigate this variance.
3. **Data-Driven Decisions**: Utilize this insight to refine forecasting and inventory management practices, as differing variances could impact demand planning and resource allocation.
In summary, recognizing the significant difference in variances allows for more informed and specialized management approaches in handling Quantity and Weight.

**How can you compare the variances of the samples using an LEVENE TEST in Python?**

```python
from scipy.stats import levene

# Sample data: replace these arrays with your actual data
group1 =  df['Quantity']
group2 =  df['Value']
group3 =  df['Weight']

# Perform Levene's test
stat, p_value = levene(group1, group2, group3)

print(f"Levene's test statistic: {stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

### Analysis:
The Levene's test statistic of 658.13 and an extremely low p-value (approximately 0) indicate strong evidence against the null hypothesis, confirming that the variances of **Weight**, **Value**, and **Quantity** are significantly different. This suggests that the variability in these three product attributes does not behave uniformly across the dataset.

**How can you compare the variances of the samples using an Bartlett's test in Python?**

```python
# Perform Bartlett's test
statistic, p_value = stats.bartlett(group1, group2, group3)

print(f"Bartlett's test statistic: {statistic}")
print(f"P-value: {p_value}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

### Analysis:
The Bartlett's test statistic of 1044.28 and an extremely low p-value (approximately 0) indicate strong evidence against the null hypothesis, confirming that the variances among the groups being compared are significantly different. **TEST OF PROPORTION USING Z TEST**

```python
# Define the proportion of interest
proportion_of_interest = 0.5  # This is an example proportion

# Calculate observed proportion
successes = df['Quantity'] > df['Quantity'].median()  # Example condition
observed_proportion = successes.mean()

# Total number of observations
n = len(df)

# Standard error calculation
standard_error = (proportion_of_interest * (1 - proportion_of_interest) / n) ** 0.5

# Z statistic calculation
z_statistic = (observed_proportion - proportion_of_interest) / standard_error

# p-value calculation
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f'Z-Statistic: {z_statistic}')
print(f'P-Value: {p_value}')

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print('Reject the null hypothesis: Proportions are significantly different.')
else:
    print('Fail to reject the null hypothesis: Proportions are not significantly different.')
```

### Analysis:
Since the p-value (0.982) is much greater than the typical significance level (e.g., 0.05), you fail to reject the null hypothesis. This means there is not enough evidence to say that the proportion of successes in your sample is significantly different from the proportion of interest (0.5).
